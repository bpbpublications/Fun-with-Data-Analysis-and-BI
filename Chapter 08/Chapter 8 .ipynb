# Data Preprocessing

```python
# Let's focus on numerical columns for clustering
numeric_columns = ['Current Price', 'Price Change', 'Volatility', 'ROE', 'Cash Ratio',
                   'Net Cash Flow', 'Net Income', 'Earnings Per Share',
                   'Estimated Shares Outstanding', 'P/E Ratio', 'P/B Ratio']
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[numeric_columns])
```

# Determining Number of Clusters 

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Select only numeric columns
numeric_data = data[numeric_columns]

# Determine optimal number of clusters using Elbow Method and Silhouette Score
inertia_values = []
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
    kmeans.fit(numeric_data)
    inertia_values.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(numeric_data, kmeans.labels_))

# Plot Elbow Method and Silhouette Score graphs
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(range(2, 11), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')

plt.subplot(1, 2, 2)
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score')
plt.tight_layout()
```

# Clustering Analysis 

```python
from sklearn.cluster import KMeans

# Perform K-means clustering
num_clusters = 5 # By using elbow method
kmeans = KMeans(n_clusters=num_clusters, random_state=0, n_init = 10)
cluster_labels = kmeans.fit_predict(numeric_data)
data['Cluster Labels'] = cluster_labels
```

## Interpretation of each cluster 

```python
# Interpret and analyze the clusters
cluster_summary = data.groupby('Cluster Labels')[numeric_columns].mean()

# Print cluster summary
print("Cluster Summary:")
print(cluster_summary)
```

```python
# Group the data by Cluster Labels
cluster_grouped = data.groupby('Cluster Labels')

# Print the Security, GICS Sector, and GICS Sub Industry for each cluster
for cluster_label, cluster_data in cluster_grouped:
    print(f"Cluster {cluster_label}:")
    print(cluster_data[['Security', 'GICS Sector', 'GICS Sub Industry']])
    print()
```

## Comparative Analysis 

```python
# Calculate average values for each numeric feature across all clusters
average_values = data[numeric_columns].mean()

# Print header for the comparative analysis
print("Comparative Analysis:")
print("-" * 40)

# Iterate through each cluster
for cluster_label, cluster_data in cluster_grouped:
    print(f"Cluster {cluster_label}:")
    
    # Calculate differences from average values for each numeric feature
    diff_from_average = cluster_data[numeric_columns].mean() - average_values
    
    # Print differences for each feature
    print("Differences from Average Values:")
    print(diff_from_average)
    
    print("-" * 40)

# Add a line break between clusters
print()
```

## Investment Insights 

```python
for cluster_label, cluster_data in cluster_grouped:
    print(f"Investment Insights for Cluster {cluster_label}:")
    
    # Analyze the key findings for this cluster
    cluster_summary = cluster_data[numeric_columns].mean()
    
    # Print insights about the cluster
    print("Cluster Summary:")
    print(cluster_summary)
    
    # Investment Insights
    if cluster_summary['Volatility'] < 2.0:
        print("This cluster consists of relatively stable stocks with low volatility, making it suitable for conservative investors.")
    elif cluster_summary['Volatility'] >= 2.0 and cluster_summary['Volatility'] < 4.0:
        print("This cluster includes stocks with moderate volatility, which may appeal to investors seeking a balance between risk and return.")
    else:
        print("Investors considering this cluster should be prepared for higher volatility, potentially leading to higher returns but also greater risk.")
    
    print()  # Add a line break between clusters
```
